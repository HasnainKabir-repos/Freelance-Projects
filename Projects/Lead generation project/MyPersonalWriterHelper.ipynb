```python
import pandas as pd

# Define the path to the input Excel file
input_excel_file = 'Oklahoma_and_Solar.xlsx'

# Read data from the input Excel file
df = pd.read_excel(input_excel_file)

# Create a new DataFrame to store the modified data
new_df = pd.DataFrame(columns=['City / State','Business', 'Website', 'Phone'])

# Iterate over each row in the input data
for index, row in df.iterrows():
    # Retrieve the required values from the input row
    name = row['name']
    city = row['city']
    region = row['region']
    website = row['website']
    phone = row['phone']
    
    # Create a new row for the modified data
    new_row = {
        'City / State': f'{city}, {region}',
        'Business': name,
        'Website': website,
        'Phone': phone
    }
    
    # Append the new row to the DataFrame
    new_df = new_df.append(new_row, ignore_index=True)

# Define the path to the output Excel file
output_excel_file = 'Modified_Oklahoma_and_Solar.xlsx'

# Write the modified data to the output Excel file
new_df.to_excel(output_excel_file, index=False)

print("Data has been written to the output Excel file successfully!")
```

```python


# Define the path to the input Excel file
input_excel_file = 'Modified_Oklahoma_and_Solar.xlsx'

# Read data from the input Excel file
df = pd.read_excel(input_excel_file)

# Remove duplicate rows based on 'Business' column
df = df.drop_duplicates(subset='Business', keep='first')

# Define the path to the output Excel file
output_excel_file = 'Modified_Oklahoma_and_Solar_NoDuplicates.xlsx'

# Save the modified data to the output Excel file
df.to_excel(output_excel_file, index=False)

print("Duplicate rows have been removed and data has been saved to the output Excel file successfully!")
```
