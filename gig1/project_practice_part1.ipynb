```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
import pandas as pd
```

```python
df = pd.DataFrame(columns=['description', 'price', 'country', 'supplier_name', 'supplier_link'])
```

### Since supplier name is not available on the immediate page, it makes the program a lot slower as we have to navigae through the page and that takes time

```python
#Load selenium driver

#Disable popups
chrome_options = webdriver.ChromeOptions()
prefs = {"profile.default_content_setting_values.notifications" : 2}
chrome_options.add_experimental_option("prefs",prefs)
driver = webdriver.Chrome(chrome_options=chrome_options)


driver.get("https://www.alibaba.com/trade/search?fsb=y&IndexArea=product_en&CatId=&SearchText=paracord&viewtype=&tab=")

for i in range(17,23):

    results = WebDriverWait(driver, 20).until(
        EC.presence_of_all_elements_located((By.XPATH, "//div[@data-content='productItem']")))

    for r in results:
        paragraph = r.find_element(By.XPATH, ".//h2[@data-e2e-name='title']")

        description = paragraph.get_attribute("title")
        
        try:
            price = r.find_element(By.XPATH, ".//p[starts-with(@data-e2e-name, 'price')]").get_attribute("title")
        except:
            price = r.find_element(By.XPATH, ".//span[starts-with(@class, 'elements-offer-price-')]").text
        
        country = r.find_element(By.XPATH, ".//span[starts-with(@class, 'seller-tag__country')]").get_attribute("title")

        supplier_link = r.find_element(By.XPATH, ".//a[@flasher-type='supplierName']").get_attribute("href")
        
        action = ActionChains(driver)
        link = r.find_element(By.XPATH, ".//a[@flasher-type='supplierName']")
        action.move_to_element(link).perform()
        
        try:
            WebDriverWait(driver, 7).until(
                EC.presence_of_element_located((By.XPATH, "//a[@class='supplier-tag-popup__content_href']"))).is_displayed()
        
            supplier = r.find_element(By.XPATH, "//a[@class='supplier-tag-popup__content_href']").text
        except:
            supplier = ""
        
        action.move_by_offset( xoffset=-100, yoffset=0).perform()

        new_row = {'description': description, 'price':price, 'country':country, 'supplier_name':supplier, 'supplier_link': supplier_link}
        df = df.append(new_row, ignore_index=True)

    next_page = driver.find_element(By.XPATH, "//a[@data-params='floorName=next&filterValue="+str(i)+"']")
    next_page.click()
```
